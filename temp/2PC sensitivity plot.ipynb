## 2PC sensitivity plot

```python
%matplotlib inline
import pyfits, pickle
os.chdir(os.path.expandvars('$HOME/guburnett/analysis/pulsars'))
from uw.like2.pub import healpix_map as hm; reload(hm) 
```

```python

tbl = np.array(pyfits.open(os.path.expandvars('$HOME/guburnett/skymodels/3years/uw10/Pulsar_limits_fix2.fits'))[1].data,float)
tbl *= 3.25e-12
tbl[np.isinf(tbl)]=0
pl = hm.HParray('pulsar_limits',tbl )
from matplotlib.colors import LogNorm
v = np.array([1,2,5,10,20])*1e-12
ait1=pl.plot(norm=LogNorm(vmin=v[0], vmax=v[-1]))
cb = ait1.colorbar
cb.set_ticks(v)
cb.set_ticklabels(['%d' %x for x in v/v[0]])
cb.set_label(r'$\mathrm{Energy\ flux\ [10^{-12} erg \ s^{-1} cm^{-2}]}$')
```
