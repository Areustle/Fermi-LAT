```python
from uw.like2 import localization, plotting; reload(localization)

def plot_tsmap(roi, source_name=None):
    source = roi.sources.find_source(source_name)
    plot_kw = dict(size=0.25, pixelsize=0.25/15, outdir=None, assoc=None, galmap=None, notitle=True,nolegend=True)
    with roi.tsmap_view(source.name) as tsm:
        loc =localization.Localization(tsm)
        tsize = source.ellipse[2]*15. # scale according to major axis s
        plot_kw.update(size=tsize, pixelsize=tsize/15.)
        tsp = plotting.tsmap.plot(loc, **plot_kw)
        return tsp
tsp= plot_tsmap(r);
```
