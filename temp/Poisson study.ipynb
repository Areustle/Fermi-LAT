##Poisson Study
Examine our Poisson-like function

$w(s;e_0,e,b) = e(s_0+b)\ \log(e(s+b)) - es$ + const

as implemented by uw.like2.loglikelihood.Poisson. Here

* $s_0$ flux at peak 
* $e$ Normalization factor to convert flux to equivalent counts
* $b$ Background flux, must be >0

```python
%matplotlib inline
from uw.like2.loglikelihood import Poisson, PoissonFitter
```

```python
w = Poisson([10.0, 1.0, 1.0])
print '%s, TS=%.1f' % (w, w.ts)
w
```

```python
fig, ax = plt.subplots(figsize=(5,5))
x = np.linspace(0,20)
y = w(x)
ax.plot(x,y)
```

Now try the fitter, check its plotting

```python
pf = PoissonFitter(w)
pf.plot();
```

Now check the integration of the probability

```python
print 'Total integral: %.2f' % w.cdfc(0)
```

## increase background
Actually, make the signal negative

```python
w2 = Poisson([-5, 1.0, 10.0])
print '%s, TS=%.1f' % (w2, w2.ts)
pf2 = PoissonFitter(w2)
pf2.plot();
```

```python
def test(w):
    beta = w.p[1]*w.p[2]
    print beta, w.cdfc(0), 1-1/w.cdfc(-beta)
test(w2), test(w)
```
