##Select subset of sources for moving
This reads the csv file generated by the sourceinfo analysis step.

```python
flagged = pd.read_csv('moment_localizations.csv', index_col=0)
cut = (flagged.ts>10) & (flagged.locqual<5) &(flagged.delta_ts>0.5)
subset = flagged[cut]
rois = flagged.roiname[cut]
rois = sorted(list(set( [int(r[5:]) for r in rois]))); rois
sources = sorted(list(subset.index));
print 'selected %d sources, in %d rois' % (len(sources), len(rois))
selected = flagged[cut].sort_index(by='roiname')['ts locqual delta_ts roiname'.split()]; selected
```

```python
r= process.Process('.', finish=False, update_positions_flag=True)
```

```python
for roi in rois[1:]:
    r.setup_roi(roi)
    r.process()
```

```python
for roi in rois[1:]:
    print roi, '='*50
    r.setup_roi(roi)
    for sourcename in sources:
        if selected.ix[sourcename].roiname=='HP12_%04d'%roi:
            r.localize(sourcename)
    r.process()
```
