```python
%matplotlib inline
os.chdir(os.path.expandvars('$FERMI/skymodels/P301_6years/uw969'))
import glob, pickle
ff = sorted(glob.glob('pickle/*.pickle')); 
norms = []
for f in ff:
    t = pickle.load(open(f,'rw')); 
    iso = t['diffuse_names'].index('isotrop')
    m = t['diffuse'][iso]
    norms.append(m[0])
iso_norms = np.array(norms)
```

```python
print 'min, mean, max: %.3f, %.3f, %.3f' % (iso_norms.min(), iso_norms.mean(), iso_norms.max())
pickle.dump(iso_norms, open('isotropic_norms.pickle', 'w'))
```

```python
vin='uw960'; vout='uw963'
ra = (1./iso_norms).clip(0.25,4.0)
for et in ('front', 'back'):
    infile = os.path.expandvars('$FERMI/diffuse/isotropic_correction_%s_%s.csv'% (et, vin))
    assert os.path.exists(infile), 'File %s not found' %infile
    cv = pd.read_csv(infile, index_col=0)
    print 'read input iso file %s' % infile
    # divide input corrections by current norms and write out
    for i in range(8):
        cv.ix[i] /= ra[i]
    outfile = os.path.expandvars('$FERMI/diffuse/isotropic_correction_%s_%s.csv'% (et,vout))
    cv.to_csv(outfile)
    print 'wrote new iso diffuse correction file %s' % outfile

```
