```python
from skymaps import SkyDir
def insert_weak_pulsar(r, roi, pname, pdir):    
    r.setup_roi(roi); 
    r.add_source(name=pname, skydir=pdir, model='PLSuperExpCutoff(1e-12, 1.5, 2000., 1.0)')
    r.freeze('b', pname)
    r.freeze('Index', pname)
    r.freeze('Cutoff', pname)
    r.fit(pname, tolerance=0)
    r.process()
```

```python
rtbl = """J0631+0646    97.968     6.771 867""".split('\n')


tbl = [x.split() for x in rtbl] ; tbl
for z in tbl:
    pname='PSR '+z[0]; roi = int(z[3]);
    pdir = SkyDir(float(z[1]), float(z[2]))
    print '='*40
    print 'ROI %d: adding %s' %(roi,  pname)
    insert_weak_pulsar(r, roi, pname, pdir)
```
