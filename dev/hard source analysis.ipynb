#Study properties of hard sources
The setup chooses a model to analyse, and loads the sourceinfo module

```python
%matplotlib inline
import os
import matplotlib.pylab as plt
import numpy as np
import pandas as pd
os.chdir(os.path.expandvars('$FERMI/skymodels/P301_5years/uw901'))
from uw.like2.analyze import sourceinfo; reload(sourceinfo)
self = sourceinfo.SourceInfo()
```

Add an element to the DataFrame object that the SourceInfo object has created from the sources.

```python
self.df['ts10k'] = [sum(sr.ts[8:]) for sr in self.df.sedrec]
```

Define a function that could be part of an analysis module, which would be a subclass of SourceInfo, then call it with self as an argument. (If it were a in the class, it would be invoked as self.Ts10k_plots())

```python
def ts10k_plots(self, tscut=25):
    fig, axx = plt.subplots(1,3, figsize=(12,5))
    plt.subplots_adjust(wspace=0.3)
    df = self.df[self.df.ts>tscut]
    
    ax = axx[0]
    ax.hist(df.ts10k.clip(0,1e3), np.logspace(1,3,21), label='TS>25')
    plt.setp(ax, xscale='log', xlabel='TS(E>10 GeV)');
    ax.legend();ax.grid()
    
    ax = axx[1]
    ax.hist(df.pindex[df.ts10k>25], np.linspace(1,3,21), label='TS10K>25')
    plt.setp(ax, xlabel='photon index')
    ax.grid(); ax.legend()
    
    ax=axx[2]
    cut =df.ts10k>10
    ax.plot(df.pindex[cut], df.ts10k[cut], '.')
    plt.setp(ax, xlim=(1,3),xlabel='photon index',
             yscale='log', ylim=(10,1e3), ylabel='TS10K')
    ax.grid()
    
    return fig
ts10k_plots(self);
```
