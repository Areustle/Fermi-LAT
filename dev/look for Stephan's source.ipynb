```python
%matplotlib inline
os.chdir(os.path.expandvars('$FERMI/skymodels/P301_5years/uw906'))
from uw.like2 import main; reload(main)
```

```python
sourcelist=dict(
FORNAX= (54.63,-35.45),
A1060= (159.2, -27.53),
A3526= (192.06, -41.35),# (192.2, -41.31), # —> This one you’ve already looked at. 
COMA=  (195.0,  27.98),
A1367= (176.1,  19.84),
)
```

```python
def analyze_cluster(name): 
    def print_ts():
        print 'TS: %.1f' % r.TS(name)
    loc = sourcelist[name]
    r = main.ROI('.', loc)
    r.print_summary(); r.summarize()
    s = r.add_source(name=name, skydir=loc, model='PLSuperExpCutoff(1e-11, 0, 2e3, 1.)')
    r.freeze('b', name, 1.0)
    r.freeze('Cutoff', name, 2000.)
    r.fit('_Norm', tolerance=0) #, setpars={10:-13})
    print_ts()
    r.fit(name, tolerance=0); print_ts()
    r.localize(update=True)
    r.fit(name, tolerance=0); print_ts()
    sed_fig=r.plot_sed(butterfly=False);
    ax=sed_fig.get_axes()[0]
    ax.set_ylim(0.05,10); 
    r.plot_tsmap();
```

```python
analyze_cluster('A3526')
```
