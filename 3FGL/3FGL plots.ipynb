```python
def copyit(infolder, filename):
    infilename =os.path.join(infolder, filename) 
    infile = open(os.path.join(infolder, filename))
    with open('lat3fg/'+filename, 'w') as outfile:
        outfile.write(infile.read())
    return Image(filename=infilename)
copyit('sedfig','PSR_J1459-6053_sed.png')
```
