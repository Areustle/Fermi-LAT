```python
def convert_to_pulsar(r, pname, ra, dec, sname):
    pdir = SkyDir(float(ra),float(dec))
    pname = 'PSR '+pname
    roi = Band(12).index(pdir)
    print '------ROI #',roi
    r.setup_roi(roi); 
    s =r.sources.find_source(sname); s.name=pname
    sdir=s.skydir
    diff = np.degrees(pdir.difference(sdir))
    print 'difference:  %.2f deg' % diff
    assert diff<0.2, 'not close enough: %s,%s' % (sname,pname)
    r.sources.set_model( 'PLSuperExpCutoff(1e-12, 1.5, 2000., 1.0)', pname)
    r.freeze('b')
    r.fit(pname)
    r.localize(update=True)
    r.process()
```
